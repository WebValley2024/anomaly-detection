```python
#show the data
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import os


def create_dir(name):
    if os.path.exists(name) == False:
        os.mkdir(name)
    else:
        os.system(f"rm -r {name}")
        os.mkdir(name)

create_dir("output/img")
create_dir("output/img/HEPP_X")
create_dir("output/img/HEPP_L")
create_dir("output/img/HEPP_H")
create_dir("output/img/EFD_ULF")
create_dir("output/img/LAP")


for l in ["HEPP_X","HEPP_L","HEPP_H","EFD_ULF","LAP"]:
        
        dirpt = os.listdir("output/"+l)
        
        for i in dirpt:

            arr = xr.open_zarr(f"output/{l}/{i}")
            
            timeser = False

            if False:

                fig, ax = plt.subplots(1,2,figsize=(15,5))
                fig.suptitle(i)
                ax[0].scatter(arr["GEO_LON"],arr["GEO_LAT"],c=arr["Block_x"],cmap="viridis")
                ax[1].scatter(arr["GEO_LON"],arr["GEO_LAT"],c=arr["Block_y"],cmap="viridis")
                plt.savefig(f"output/img/{l}/{i}.png")
                plt.close()

            else:

                if l == "HEPP_L":
                    
                    fig, ax = plt.subplots(9,1,figsize=(90,20))
                    
                    for n in range(9):
                        if timeser:
                            ax[n].scatter(arr["TIME"].values,arr[f'A411_0_{n}'][:].values)
                        else:
                            ax[n].plot(range(len(arr[f'A411_0_{n}'][:].values)),arr[f'A411_0_{n}'][:].values)
                    
                    plt.savefig(f"output/img/{l}/{i}.png")
                    plt.close()

                elif l == "HEPP_X":
                    fig, ax = plt.subplots(figsize=(90,20))
                    
                    if timeser:
                        plt.scatter(arr["TIME"].values,arr[f'XrayRate'].values)
                    else:
                        plt.plot(range(len(arr[f'XrayRate'].values)),arr[f'XrayRate'].values)
                    
                    
                    plt.savefig(f"output/img/{l}/{i}.png")
                    plt.close()

                elif l == "HEPP_H":
                    fig, ax = plt.subplots(2,1,figsize=(90,20))
                    
                    if timeser:
                         
                        ax[0].scatter(arr["TIME"].values,arr[f'Count_Electron'].values)
                        ax[1].scatter(arr["TIME"].values,arr[f'Count_Proton'].values)
                    
                    else:
                        
                        ax[0].plot(range(len(arr[f'Count_Electron'].values)),arr[f'Count_Electron'].values)
                        ax[1].plot(range(len(arr[f'Count_Proton'].values)),arr[f'Count_Proton'].values)
                    
                    plt.savefig(f"output/img/{l}/{i}.png")
                    plt.close()

                elif l == "EFD_ULF":
                    fig, ax = plt.subplots(3,1,figsize=(90,20))
                    
                    for n in range(3):
                        
                        if timeser:
                         
                            ax[n].scatter(arr["TIME"].values,arr[f'A11{n+1}_W'][:].values)
        
                        else:
                            
                            ax[n].plot(range(len(arr[f'A11{n+1}_W'][:].values)),arr[f'A11{n+1}_W'][:].values)
                    
                    plt.savefig(f"output/img/{l}/{i}.png")
                    plt.close()
                
                elif l == "LAP":
                    fig, ax = plt.subplots(2,1,figsize=(90,18))
                    
                    for n in range(2):
                        
                        if timeser:
                         
                            ax[n].scatter(arr["TIME"].values,arr[f'A3{n+1}1'][:].values)
        
                        else:
                            
                            ax[n].plot(range(len(arr[f'A3{n+1}1'][:].values)),arr[f'A3{n+1}1'][:].values)
                    
                    plt.savefig(f"output/img/{l}/{i}.png")
                    plt.close()

```
